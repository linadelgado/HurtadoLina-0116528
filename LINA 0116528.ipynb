```python
6.
lista=[]
def fib(n):
    a,b = 0,1 
    while a<n:
        lista.append(a)
        a,b= b, a+b 
    return lista
num=input('Ingrese un número:')
suma=0
for i in num:
    suma=suma+int(i)
m=1000
fibo=fib(m)
bandera=0
for i in fibo:
    if i ==suma:
        bandera=1
if bandera ==1:
    print('La suma del número se encuentra en la sucesión de fibonacci ')
else:
    print('La suma del número no se encuentra en la sucesión de fibonacci ')
```
